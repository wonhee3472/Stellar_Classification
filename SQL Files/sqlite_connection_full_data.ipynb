```python
# Create db file
from pathlib import Path
Path("stellar_class_db").touch()
```

```python
import sqlite3

# Create db connection and cursor to execute queries

conn = sqlite3.connect("stellar_class_db")
c = conn.cursor()
```

```python
# Create empty stellar table

c.execute('''CREATE TABLE stellar (obj_ID float, alpha float, delta float, u float, g float, r float, i float, z float, run_ID int, rerun_ID int, cam_col int, field_ID int, spec_obj_ID float, class text, redshift float, plate int, mjd int, fiber_ID int )''')
```

```python
# Load csv into pandas df

stellar = pd.read_csv("Resources/star_classification.csv")

# write data to sqlite table stellar
stellar.to_sql("stellar", conn, if_exists="append", index=False)
```
